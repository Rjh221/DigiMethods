## 5.1.1 Data Wrangling and Visualizing

**Exercise 5.1.1.1** Read in the dataset "nchs_death_causes.csv" as a list of lists. You will need to use the `csv` package and then will want to go through each row appending it to a list you create.

```python
import csv 

f = open ("nchs_death_causes.csv", "r")
reader = csv.reader(f)


better_list_of_lists = []
Nebraska_suicide = []
for row in reader:
    better_list_of_lists.append(row)
f.close()
```

**Exercise 5.1.1.2** Make a list that is a subset of the data, containing only one cause of death and one state.

```python

for element in better_list_of_lists:
    if element[2] == "Suicide" and element[3] == "Nebraska": 
        Nebraska_suicide.append(element)

Nebraska_suicide.reverse()
```

**Exercise 5.1.1.3** Plot a line graph that shows the development of the death rate over time.

```python
import matplotlib.pyplot as plt
import os


values = [element[0] for element in Nebraska_suicide]
labels = [int(element[4]) for element in Nebraska_suicide]
plt.figure(figsize=(10, 7), dpi= 120)

plt.plot(values, labels)

plt.title("Graph to show number of suicides in Nebraska by year")
plt.xlabel ("Year")
plt.ylabel ("Suicides by 100,000")
plt.xticks (range(0, len(labels)), rotation=90)


file_dir = os.getcwd()

if '/' in file_dir:
    print("I am a Mac.")
    plt.savefig(file_dir + "/Suicide_in_nebraska.png")
```

**Exercise 5.1.1.5** Make a list that is a subset of the data, containing _two_ causes of death that might be interesting to compare for one state.

```python
double_list = []
double_list.append(Nebraska_suicide)

unintentional_list = []
for rows in better_list_of_lists: 
    if rows [2] == "Unintentional injuries" and rows[3]=="Nebraska":
        unintentional_list.append(rows)
double_list.append(unintentional_list)


```

**Exercise 5.1.1.6** Create a _self-contained_ scatter plot that compares the two causes of death in the state.

```python
import matplotlib.pyplot as plt

values_suicide = [int(item[4]) for item in double_list[0]]
values_u = [int(item[4]) for item in double_list[1]]

plt.figure(figsize=(10, 7), dpi= 120)

plt.xlabel ("No. of suicides")
plt.ylabel ("No. of unintentional accidents resulting in death")
plt.title ("Scatterplot comparing number of suicides and unintentional accidents resulting in death in Nebraska")
plt.plot(values_suicide, values_u, 'o')
```
